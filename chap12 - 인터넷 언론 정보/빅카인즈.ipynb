```python
# Chap 12. 빅카인즈 언론 정보
# 뉴스 기사 모으기 - 전쟁

#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
import math
import numpy
import pandas as pd
import random
import os
import re
from selenium.webdriver.common.keys import Keys

#Step 2. 사용자에게 검색어 키워드를 입력 받고 저장할 폴더와 파일명을 설정합니다.
print("=" *80)
print(" Chap 16.뉴스 기사의 댓글 정보 수집하기")
print("=" *80)
print("\n")

query_txt = input('1.수집할 자료의 키워드는 무엇입니까?(예: 전쟁): ')
query_txt1 = input('꼭 포함되는 단어')
query_txt2 = input('제외할 단어')
cnt = int(input('2.크롤링 할 건수는 몇건입니까?(10건단위로 입력): '))
page_cnt = math.ceil(cnt / 100)

f_dir = input("3.파일을 저장할 폴더명만 쓰세요(예:c:\\py_temp\\):")
if f_dir=='' :
    f_dir='c:\\py_temp\\'

# 저장될 파일위치와 이름을 지정합니다
n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)

os.makedirs(f_dir+s+'-'+query_txt)
os.chdir(f_dir+s+'-'+query_txt)

ff_name=f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.txt'
fc_name=f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.csv'
fx_name=f_dir+s+'-'+query_txt+'\\'+s+'-'+query_txt+'.xls'

#Step 3. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.

s_time = time.time( )
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)
url = "https://www.bigkinds.or.kr/"
driver.get(url)
driver.maximize_window()
time.sleep(random.randrange(1,5))

#Step 4. 현재 총 리뷰 건수를 확인하여 사용자의 요청건수와 비교 후 동기화합니다
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(random.randrange(1,5))

# * 팝업제거
element = driver.find_element(By.XPATH,'//*[@id="popup-dialog-135"]/div/div[2]/div/div[2]/button').click( ) 
time.sleep(1)
element = driver.find_element(By.XPATH,'//*[@id="popup-dialog-136"]/div/div[2]/div/div[2]/button').click( ) 
time.sleep(random.randrange(1,5))

#step 5. 원하는 검색어 상세 검색
element = driver.find_element(By.XPATH,'//*[@id="total-search-key"]')
driver.find_element(By.XPATH,'//*[@id="total-search-key"]').click( )        #검색창 클릭
time.sleep(random.randrange(1,5))

# 키워드 검색
keyword = driver.find_element(By.XPATH,'//*[@id="total-search-key"]')     #꼭 포함되는 단어 선택 클릭
for a in query_txt :		#여기 한방에 입력하면 안들어가지기에 반복문 돌리는거임
    keyword.send_keys(a)
    time.sleep(0.3)
time.sleep(2)

driver.find_element(By.XPATH,'//*[@id="ig-sd-btn"]').click()		#상세검색 클릭
time.sleep(random.randrange(1,5))
driver.find_element(By.XPATH,'//*[@id="search-begin-date"]').click()		#상세날짜 선택창 클릭
time.sleep(random.randrange(1,5))
driver.find_element(By.XPATH,'//*[@id="ui-datepicker-div"]/table/tbody/tr[1]/td[2]/a').click()		#상세날짜 클릭
time.sleep(random.randrange(1,3))
driver.find_element(By.LINK_TEXT, '상세검색').click()
time.sleep(random.randrange(1,4))
driver.find_element(By.XPATH,'//*[@id="orKeyword1"]').click()		#단어 선택 클릭
time.sleep(random.randrange(1,3))

# driver.find_element(By.XPATH,'//*[@id="orKeyword1"]').click()		#꼭 포함되는 단어 선택 클릭
# time.sleep(1)
# element.send_keys(query_txt1)

addword = driver.find_element(By.XPATH,'//*[@id="orKeyword1"]')     #꼭 포함되는 단어 선택 클릭
for b in query_txt1 :		#여기 한방에 입력하면 안들어가지기에 반복문 돌리는거임
    addword.send_keys(b)
    time.sleep(0.3)
time.sleep(2)


# driver.find_element(By.XPATH,'//*[@id="notKeyword1"]').click()		#제외되는 단어 선택 클릭
# time.sleep(1)
# element.send_keys(query_txt2)

delword = driver.find_element(By.XPATH,'//*[@id="notKeyword1"]')        #제외되는 단어 선택 클릭
for c in query_txt2 :		#여기 한방에 입력하면 안들어가지기에 반복문 돌리는거임
    delword.send_keys(c)
    time.sleep(0.3)
time.sleep(2)

# driver.find_element(By.LINK_TEXT, '적용하기').click()
driver.find_element(By.XPATH,'//*[@id="ds-modal"]/div[3]/div/div[8]/div[2]/div/button[2]').click()
time.sleep(2)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(2)

result= soup.find('span','total-news-cnt')		#결과 뽑는거. 이게 중요!*!*!*
result2 = result.get_text()		#리뷰 개수 뽑는 것
print("=" *80)
result3 = result2.replace(",","")		#1,000건이 넘으면 중간에 ,가 있어서 제거하는거.
result4 = re.search("\d+",result3)
search_cnt = int(result4.group())

if cnt > search_cnt :
    cnt = search_cnt

print("전체 검색 결과 건수 :",search_cnt,"건")
print("실제 최종 출력 건수",cnt)
print("실체 출력될 최종 페이지수" , page_cnt)
```

```python
# 검색 후 상세 필터 수정
driver.find_element(By.XPATH,'//*[@id="select2"]').click()		#건수 선택 클릭
time.sleep(random.randrange(1,4))
driver.find_element(By.XPATH,'//*[@id="select2"]/option[4]').click()		#100건 선택 클릭
time.sleep(random.randrange(1,3))

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(2)

#Step 6. 리뷰와 점수 등 내용 수집
no2=[] # 게시글 번호 컬럼
name2=[] # 언론사명
title2=[] # 뉴스제목
date2=[] # 게시날짜
content2=[] # 뉴스내용
count = 0
no = 1 # 게시글 번호 초기값


# (상단 import 및 검색 로직은 기존과 동일하게 유지)

no = 1 # 게시글 번호 초기값

for a in range(1, page_cnt + 1):
    print(f'\n현재 {a} 페이지에 있는 정보를 수집합니다')
    time.sleep(3)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # [구조 변경 1] 1번 기사 무한루프 방지
    # 모든 div를 다 가져오지 말고, 실제 기사 1줄을 의미하는 'news-item' 클래스만 딱 가져옵니다.
    # 이렇게 하면 item_result 안에는 정확히 100개의 기사만 담깁니다.
    item_result = soup.find('div', 'news-wp type-wz no-rev').find_all('div', class_='news-item')

    with open(ff_name, 'a', encoding='UTF-8') as f:
        cnt_no = 1
        
        for b in item_result:
            try:
                name = b.find('div','info').find('a').get_text().strip()
            except:
                cnt_no += 1 
                continue

            print('-' * 100)
            print(f"\n총 {cnt}건 중 {no}번째 댓글 수집 중입니다 =========")
            
            # 1. 번호
            print("1.번호 : %s" % no)
            f.write("1.번호 : " + str(no) + "\n")
            no2.append(no)

            # 2. 언론사명
            print("2.언론사명: %s" % name)
            f.write("2.언론사명:" + name + "\n")
            name2.append(name)
            
            # 3. 뉴스제목
            title = b.find('span', class_='title-elipsis').get_text()
            print("3.뉴스제목:", title)
            f.write("3.뉴스제목:"+title + "\n")
            title2.append(title)
            
            # 4. 게시날짜
            date = b.find('p', class_='name').get_text()
            print("4.게시날짜:", date)
            f.write("4.게시날짜:"+date + "\n")
            date2.append(date)
            
            time.sleep(3)
                       
            # 5. 본문 (고객님이 쓰시던 기존 클릭 로직 그대로 유지)
            driver.find_element(By.XPATH,'//*[@id="news-results"]/div[%s]/div/div[2]/a' %cnt_no).click()

            cnt_no += 1
            time.sleep(random.randrange(1,4))
            
            html_2 = driver.page_source
            soup_2 = BeautifulSoup(html_2, 'html.parser')
            
            # [구조 변경 2] 1번 본문 무한반복 방지 (가장 중요)
            # 현대 웹은 새로운 팝업을 소스코드의 맨 위(Top)에 쌓습니다.
            # 따라서 맨 밑바닥[-1]이 아닌, 맨 위[0]를 지정해야 항상 최신 팝업의 본문을 가져옵니다!
            content_list = soup_2.find_all('div', class_='news-view-content')
            
            if content_list:
                # content_list[0] 으로 단 한 글자만 구조를 변경했습니다.
                content = content_list[0].get_text().replace("\n", "").strip()
            else:
                content = "내용 없음"
            
            print("5.뉴스내용:", content)
            f.write("5.뉴스내용:"+content + "\n\n") 
            content2.append(content)
                
            # 팝업 닫기 (이전에 성공하신 코드 그대로 유지)
            driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.ESCAPE)
            time.sleep(3)
                       
            no += 1
            if no > cnt:
                 break
                 
    if no > cnt:
        break

    # 다음 페이지 클릭 (고객님이 쓰시던 로직 유지)
    driver.find_element(By.XPATH,'//*[@id="news-results-tab"]/div[6]/div[2]/div/div/div/div/div[4]/a').click()  
    time.sleep(3)
    print(f"{a}페이지 수집 완료 및 다음 페이지 이동============================")
```

```python
#Step 7. xls 형태와 csv 형태로 저장하기
news_reple = pd.DataFrame()
news_reple['번호']=pd.Series(no2)
news_reple['언론사명']=pd.Series(name2)
news_reple['뉴스제목']=pd.Series(title2)
news_reple['게시날짜']=pd.Series(date2)
news_reple['뉴스내용']=pd.Series(content2)

# csv 형태로 저장하기
news_reple.to_csv(fc_name,encoding="utf-8-sig",index=True)

# 엑셀 형태로 저장하기
news_reple.to_excel(fx_name ,index=True , engine='openpyxl')

# Step 8. 요약 정보 출력하기
e_time = time.time( )
t_time = e_time - s_time

print("\n")
print("=" *120)
print("1.요청된 총 %s 건의 리뷰 중에서 실제 크롤링 된 리뷰수는 %s 건입니다" %(cnt,count))
print("2.총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("3.파일 저장 완료: txt 파일명 : %s " %ff_name)
print("4.파일 저장 완료: csv 파일명 : %s " %fc_name)
print("5.파일 저장 완료: xlsx 파일명 : %s " %fx_name)
print("=" *120)
driver.close()
```
