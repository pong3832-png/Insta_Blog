```python
import os
import re
import zipfile
import requests
import subprocess

# --------------------------------------------------
# 1️⃣ Chrome 버전 확인
# --------------------------------------------------

def get_chrome_version():
    try:
        output = subprocess.check_output(
            r'reg query "HKEY_CURRENT_USER\Software\Google\Chrome\BLBeacon" /v version',
            shell=True
        ).decode()

        version = re.search(r"\d+\.\d+\.\d+\.\d+", output).group()
        return version
    except Exception as e:
        print("Chrome 버전 확인 실패:", e)
        return None


# --------------------------------------------------
# 2️⃣ ChromeDriver 다운로드 URL 찾기
# (Chrome 115 이상 공식 방식)
# --------------------------------------------------

def get_driver_download_url(chrome_version):

    api_url = "https://googlechromelabs.github.io/chrome-for-testing/known-good-versions-with-downloads.json"
    data = requests.get(api_url).json()

    for v in data["versions"]:
        if v["version"] == chrome_version:
            for item in v["downloads"]["chromedriver"]:
                if item["platform"] == "win64":
                    return item["url"]

    # 정확히 일치 버전이 없으면 메이저 버전 매칭
    major = chrome_version.split(".")[0]

    for v in data["versions"]:
        if v["version"].startswith(major):
            for item in v["downloads"]["chromedriver"]:
                if item["platform"] == "win64":
                    return item["url"]

    return None


# --------------------------------------------------
# 3️⃣ 다운로드 + 압축 해제
# --------------------------------------------------

def download_driver(url):
    zip_name = "chromedriver.zip"
    extract_dir = os.path.abspath("chromedriver")

    print("📥 ChromeDriver 다운로드 중...")
    r = requests.get(url)

    with open(zip_name, "wb") as f:
        f.write(r.content)

    with zipfile.ZipFile(zip_name, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)

    os.remove(zip_name)

    # chromedriver.exe 경로 찾기
    for root, dirs, files in os.walk(extract_dir):
        for file in files:
            if file == "chromedriver.exe":
                return os.path.join(root, file)

    return None


# --------------------------------------------------
# 4️⃣ Driver 버전 확인
# --------------------------------------------------

def get_driver_version(driver_path):
    output = subprocess.check_output([driver_path, "--version"]).decode()
    version = re.search(r"\d+\.\d+\.\d+\.\d+", output).group()
    return version


# --------------------------------------------------
# 5️⃣ 실행
# --------------------------------------------------

def main():

    print("==== Chrome 정보 ====")

    chrome_version = get_chrome_version()

    if not chrome_version:
        print("Chrome을 찾을 수 없습니다.")
        return

    print("Chrome 버전:", chrome_version)

    print("\n==== ChromeDriver 다운로드 ====")

    url = get_driver_download_url(chrome_version)

    if not url:
        print("일치하는 ChromeDriver를 찾지 못했습니다.")
        return

    driver_path = download_driver(url)

    print("Driver 저장 위치:", driver_path)

    driver_version = get_driver_version(driver_path)

    print("\n==== 버전 비교 ====")
    print("Chrome 버전 :", chrome_version)
    print("Driver 버전 :", driver_version)

    if chrome_version.split(".")[0] == driver_version.split(".")[0]:
        print("✅ 메이저 버전 일치")
    else:
        print("❌ 메이저 버전 불일치")


if __name__ == "__main__":
    main()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

query_txt = input('검색어 입력: ')

driver = webdriver.Chrome()
driver.get("https://www.musinsa.com")
driver.maximize_window()

wait = WebDriverWait(driver, 10)

time.sleep(2)
# 검색 버튼 클릭
search_button = wait.until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, '[data-button-id="search_window"]'))
)
search_button.click()

time.sleep(2)
# 실제 input 선택
search_input = wait.until(
    EC.element_to_be_clickable(
        (By.CSS_SELECTOR, 'input[placeholder="검색어를 입력하세요"]')
    )
)
time.sleep(1)
search_input.send_keys(query_txt)
search_input.send_keys(Keys.ENTER)
```
